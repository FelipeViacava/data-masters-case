### Colunas Duplicadas e Colunas Constantes

```python
df = train.copy()
id_target = ["TARGET", "ID"]
```

```python
# Colunas constantes
constant_cols = [
    col
    for col in df.columns
    if
        (df[col].nunique() == 1) &
        (col not in id_target)
]
```

```python
all_cols = [
    col
    for col in df.columns
    if ((col not in constant_cols) 
        & (col not in id_target))
]
keep_cols = []
drop_cols = []

for col0 in all_cols:
    for col1 in all_cols:
        if (col0 != col1):
            if (sum(df[col0] == df[col1]) == len(df)):
                if col0 not in drop_cols:
                    keep_cols.append(col0)
                if col1 not in keep_cols:
                    drop_cols.append(col1)

drop_cols += constant_cols
```

```python
with open("resources/drop_cols.txt", 'w') as f:
    for col in drop_cols:
        f.write(str(col) + "\n")
```

```python
with open("resources/drop_cols.txt") as f:
    drop_cols = [line.rstrip("\n") for line in f]
```

```python
df = df.drop(drop_cols, axis=1)
```
